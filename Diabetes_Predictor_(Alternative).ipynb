```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Load dataset
df = pd.read_csv("/content/drive/MyDrive/diabetes_prediction_dataset.csv")
df = df.head(10000)
```

```python
# Encode gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])
```

```python
# Features and target
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(probability=True)
}

scores = {}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    scores[name] = acc
```

```python
# Display all model accuracies
print("Model Accuracies:")
for name, acc in scores.items():
    print(f"{name}: {acc:.2f}")

# Select and display the best model
best_model_name = max(scores, key=scores.get)
best_model = models[best_model_name]
print(f"\nBest model: {best_model_name} with accuracy {scores[best_model_name]:.2f}")
```

```python
import gradio as gr
import numpy as np
```

```python
def predict_diabetes(gender, age, hypertension, heart_disease, bmi, hba1c, glucose):
    if gender=='Male':
        gender=1
    else:
        gender=0
    input_data = np.array([[gender, age, hypertension, heart_disease, bmi, hba1c, glucose]])
    input_scaled = scaler.transform(input_data)
    prediction = best_model.predict(input_scaled)[0]
    return "Diabetic" if prediction == 1 else "Non-Diabetic"
```

```python
interface = gr.Interface(
    fn=predict_diabetes,
    inputs=[
        gr.Dropdown(["Male", "Female"], label="Gender"),
        gr.Number(label="Age"),
        gr.Dropdown([0, 1], label="Hypertension"),
        gr.Dropdown([0, 1], label="Heart Disease"),
        gr.Number(label="BMI"),
        gr.Number(label="HbA1c Level"),
        gr.Number(label="Blood Glucose Level")
    ],
    outputs="text",
    title="Diabetes Prediction App"
)
```
